```python
df = pd.read_csv("..\data\creditcard.csv")
df = df.drop(columns = ["Time"])
```

```python
data_unseen = df.sample(frac=0.1, random_state=42)

df = df.drop(data_unseen.index)

df.head()
```

```python
data_unseen.to_csv("..\\data\\test.csv")
```

```python
data_unseen.head()
```

```python
from pycaret.classification import *
```

```python
expt2 = setup(
data = df,
target = "Class",
session_id = 42,
train_size = 0.8, 
normalize = True,
fix_imbalance = True,
log_experiment = True,
experiment_name = "Credit Card Fraud - Expt 2",
log_plots = True)
```

```python
models()
```

```python
top2 = compare_models(sort="F1", exclude=["gpc", "mlp", "qda", "lda", "nb"], fold=5, n_select=2)
```

```python
for model in top2:
    print(model)
```

```python
et = top2[0]
tuned_et = tune_model(et, optimize="F1", fold=5)
```

```python
rf = top2[1]
tuned_rf = tune_model(rf, optimize="F1", fold=5)
```

```python
plot_model(tuned_et, plot='auc')
```

```python
plot_model(tuned_et, plot="confusion_matrix")
```

```python
plot_model(tuned_rf, plot='auc')
```

```python
plot_model(tuned_rf, plot="confusion_matrix")
```

```python
predict_model(tuned_et)
```

```python
predict_model(tuned_rf)
```

```python
final_et = finalize_model(tuned_et)
```

```python
final_rf = finalize_model(tuned_rf)
```

```python
deploy_model(model = final_rf, model_name = "final_rf_deployed", platform = "aws", authentication = {'bucket' : 'experimentspycaret'})
```
